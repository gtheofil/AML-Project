```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取 CSV 文件
file_path = r"E:\MSC\Spring\AML\GestureLink\data\GZA\sensor_data1.csv"  # 替换为你的文件路径
df = pd.read_csv(file_path, header=None)

# 假设第一列是时间
time_col = df.iloc[:, 0]
data_cols = df.iloc[:, 1:2]

# 识别手势分界点（时间间隔 > 500）
breakpoints = time_col.diff() > 500
segment_times = time_col[breakpoints]

# 可视化 10 个传感器通道数据
plt.figure(figsize=(15, 10))
for col in data_cols.columns:
    plt.plot(time_col, data_cols[col], label=f'Channel {col}', alpha=0.7)

# 在手势分界点处画竖线
for seg_time in segment_times:
    plt.axvline(x=seg_time, color='red', linestyle='--', alpha=0.8)

plt.title('Sensor Data from 10 Channels with Gesture Boundaries')
plt.xlabel('Time')
plt.ylabel('Sensor Value')
plt.legend()
plt.show()
```
